# Identifying fraudulent credit card transactions using ANN model.

```python
# Importing Libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sn
import keras

np.random.seed(2)
%matplotlib inline
```

```python
# Reading the dataset
data = pd.read_csv('credit_card.csv')
print(data.shape)
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
data['NormalizedAmount']=StandardScaler().fit_transform(data['Amount'].values.reshape(-1,1))
data = data.drop(['Amount'], axis=1)
```

```python
data = data.drop(['Time'],axis=1)
data.head()
```

```python
X = data.iloc[:,data.columns!='Class']
y = data.iloc[:,data.columns=='Class']
```

```python
X.info()
```

```python
y.head()
```

## Splitting the data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
print(X_train)
print(y_train)
print(X_test)
print(y_test)
```

## Creating the ANN model

```python
# Importing the Keras libraries and package
import keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
# Initializing the ANN
classifier1 = Sequential()

# Adding the input layer and the hidden layer
classifier1.add(Dense(units=25, kernel_initializer='uniform', activation='relu', input_dim=29))

# Adding the output layer
classifier1.add(Dense(units=1, kernel_initializer='uniform', activation='sigmoid'))
```

```python
classifier1.summary()
```

```python
# Compiling the ANN
classifier1.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
# Fitting the ANN to the Training set
classifier1.fit(X_train, y_train, batch_size = 32, epochs = 10)
```

```python
# Predicting the Test set results
y_pred = classifier1.predict(X_test)
y_pred[:10]
```

```python
y_pred = (y_pred > 0.5)
y_pred[:10]
```

```python
# Calculating the accuracy
score = classifier1.evaluate(X_test, y_test)
score
```

```python
# Making the Confusion Matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```

```python
cm
```

```python
# Confusion Matrix
from sklearn.metrics import accuracy_score
cm = confusion_matrix(y_test, y_pred) # rows = truth, cols = prediction
df_cm = pd.DataFrame(cm, index = (0, 1), columns = (0, 1))
plt.figure(figsize = (10,7))
sn.set(font_scale=1.4)
sn.heatmap(df_cm, annot=True, fmt='g')
print("Test Data Accuracy: %0.4f" % accuracy_score(y_test, y_pred))
```

```python
# Classification report
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
